```python
import numpy as np
```

```python
def Creat_matrix(name = 'A', size = (2,3)):
    H, W = size[0], size[1]
    #mat = np.zeros((H,W)).astype('str')
    mat = np.frompyfunc(list, 0, 1)(np.empty(size, dtype=object))  
    for i in range(H):
        for j in range(W):
            mat[i][j] = [name+f'_{i}_{j}']
    return mat
```

```python
A = Creat_matrix()
print(A)
```

```python
B = Creat_matrix('B', (4,5))
print(B)
```

```python
C = Creat_matrix('C', (1,2))
print(C)
```

```python
def kron_alge(A, B):
    shape_A = A.shape
    shape_B = B.shape
    size = (shape_A[0] * shape_B[0], shape_A[1] * shape_B[1])
    mat = np.frompyfunc(list, 0, 1)(np.empty(size, dtype=object))  
    #print('mat', mat)
    #First, there will be A.shape[0] x A.shape[1] blocks, each block shape B[0] x B[1]
    for a0 in range(shape_A[0]):
        for a1 in range(shape_A[1]):
            #So, the starting point is a0* shape_B[0], a1 *shape_B[1]
            str0 = a0* shape_B[0]
            str1 = a1 *shape_B[1]
            
            for b0 in range(shape_B[0]):
                for b1 in range(shape_B[1]):
                    mat[str0 + b0][str1 + b1] = A[a0][a1] + B[b0][b1]
    return mat
            
```

```python
A_kron_B = kron_alge(A, B)
AkBkC = kron_alge(A_kron_B, C)
```
